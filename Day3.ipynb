```python
import numpy

with open('data/03.txt', 'r') as file:
    data = file.read()

puzzle = []

for line in data.splitlines():
    puzzle.append(list(line))

puzzle = numpy.array(puzzle)

rows, cols = puzzle.shape

def adj(x, y):
    for dx in [-1, 0, 1]:
        for dy in [-1, 0, 1]:
            cx, cy = x + dx, y + dy

            if dx == 0 and dy == 0:
                continue

            if cx < 0 or cy < 0:
                continue

            if cx > rows - 1 or cy > rows - 1:
                continue

            yield puzzle[cx, cy], cx, cy

sum = 0
num = []
found = {}
gears = numpy.zeros([rows, cols], numpy.int64)
ratios = numpy.ones([rows, cols], numpy.int64)

for i in range (rows):
    for j in range (cols):
        if puzzle[i, j].isdigit():
            num.append(puzzle[i, j])

            for n, x, y in adj(i, j):
                if n == '*':
                    found[x + y * rows] = [x, y]
                    break

        else:
            for x, y in found.values():
                gears[x, y] += 1
                ratios[x, y] *= int(''.join(num))

            num = []
            found = {}

    for x, y in found.values():
        gears[x, y] += 1
        ratios[x, y] *= int(''.join(num))

    num = []
    found = {}

for i in range (rows):
    for j in range (cols):
        if gears[i, j] == 2:
            sum += ratios[i, j]

sum
```
